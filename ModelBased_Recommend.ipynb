```
from sklearn.feature_extraction.text import TfidfVectorizer
tvf=TfidfVectorizer(min_df=3, max_features=None ,strip_accents='unicode',analyzer='word',token_pattern=r'\w{1,}',
                   ngram_range=(1,5),stop_words='english')
```

```
Index=pd.Series(posts.index,index=posts['title']).drop_duplicates()
```

```
Index
```

```
def content_Based_Recmnd(title,n,sgd=sgd):
    index=Index[title]
    sigmoid_score=list(enumerate(sgd[index]))
    sigmoid_score=sorted(sigmoid_score, key=lambda x : x[1] ,reverse=True)
    sigmoid_score=sigmoid_score[1:n+1]
    idx=[i[0] for i in sigmoid_score]
    return posts[['title','category',' post_type']].iloc[idx]
```

```
Rating_pivot = views.pivot(index = 'user_id', columns ='post_id', values = 'rating').fillna(0)
Rating_pivot
```

```
from scipy.sparse import csr_matrix
Rating_matrix=csr_matrix(Rating_pivot.values)
```

```
Ratings_mean = np.mean(Rating_matrix, axis = 1)
Ratings_nrmz = Rating_matrix - Ratings_mean.reshape(-1, 1)
```

```
Ratings_nrmz
```

```
from scipy.sparse.linalg import svds
Usr, sigma, pst = svds(Ratings_nrmz, k = 50)
```

```
sigma = np.diag(sigma)
```

```
predicted_ratings = np.dot(np.dot(Usr, sigma), pst) + Ratings_mean.reshape(-1, 1)
```

```
pred = pd.DataFrame(predicted_ratings, columns = Rating_pivot.columns)
pred.head()
```

```
Index2=pd.Series(users.index,index=users['user_id']).drop_duplicates()
```

```
Index2
```

```
def Collaborative_Recommend(userID, number):
    index=Index2[userID]
    sorted_predictions = pred.iloc[index].sort_values(ascending=False) 
    user_data= views[views.user_id == (userID)]
    user_viewed = (user_data.merge(posts,on = 'post_id').sort_values(['rating'], ascending=False).iloc[:,:-1])
    recommendations = (posts[~posts['post_id'].isin(user_viewed['post_id'])].merge(pd.DataFrame(sorted_predictions).reset_index(), how = 'left',
               left_on = 'post_id',right_on = 'post_id').rename(columns = {index: 'Predictions'}).
         sort_values('Predictions', ascending = False).
                       iloc[:number, :-2]
                      )

    return user_viewed ,recommendations
```
