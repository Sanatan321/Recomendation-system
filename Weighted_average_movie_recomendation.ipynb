```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
credits=pd.read_csv("Movie/tmdb_5000_credits.csv")
movies=pd.read_csv("Movie/tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies.head()
```

```python
print(credits.shape)
movies.shape
```

```python
credits_renamed=credits.rename(columns={"movie_id":"id"})
```

```python
credits_renamed.head()
```

```python
movies_merged=movies.merge(credits_renamed,on="id")
```

```python
movies_merged.head()
```

```python
column=["homepage","status","title_x","title_y","production_countries"]
movies_clean=movies_merged.drop(columns=column)
```

```python
movies_clean.head()
```

```python
movies_clean.info()
```

## http://trailerpark.weebly.com/imdb-rating.html?source=post_page



```python
R=movies_clean['vote_average']
v=movies_clean['vote_count']
m=movies_clean['vote_count'].quantile(0.75)
C=movies_clean['vote_average'].mean()
```

```python
v
```

```python
movies_clean['weighted_average']=(R*v+C*m)/(v+m)
```

```python
movies_clean.head()
```

```python
movies_sortedList=movies_clean.sort_values('weighted_average',ascending=False)
```

```python
movies_sortedList[['original_title','id','vote_average','vote_count','weighted_average']].head(20)
```

```python
plt.figure(figsize=(12,6))
sb.barplot(x=movies_sortedList['weighted_average'].head(10),y=movies_sortedList['original_title'].head(10),data=movies_sortedList)
```

```python
Popularity=movies_clean.sort_values('popularity',ascending= False)
plt.figure(figsize=(12,6))
sb.barplot(x=Popularity['popularity'].head(10),y=Popularity['original_title'].head(10),data=Popularity)
```

```python
from sklearn.preprocessing import MinMaxScaler
scl=MinMaxScaler()
movies_scaled=scl.fit_transform(movies_clean[['weighted_average','popularity']])
movies_normalized=pd.DataFrame(movies_scaled,columns=['weighted_average','popularity'])
```

```python
movies_normalized.head()
```

```python
movies_clean[['normalized_weighted_average','normalized_popularity']]=movies_normalized
```

```python
movies_clean.head()
```

```python
movies_clean['score']=movies_clean['normalized_popularity']*.50+movies_clean['normalized_weighted_average']*.50
```

```python
movies_clean.head()
```

```python
movies_scored=movies_clean.sort_values('score',ascending=False)
movies_scored[['original_title','id','normalized_popularity','normalized_weighted_average','score']].head(20)
```
